```python
## Tool to sort models in model dir by various parameters
## Dir name format: "{TITLE}-e{EPOCHS}-b{BATCH_SIZE}-s{SEQ_LEN}-fpp{FUTURE_PERIOD_PREDICT}-{TIME}-e{EPOCH}-vacc{VACC}-IGNORED"

import os
import json
```

```python
ls = os.listdir("models")
```

```python
formatted_ls = []

def format_name(s):
    d = {}

    ssplit = s.split("-")

    d["title"] = ssplit[0]
    d["epochs"] = int(ssplit[1][1:])
    d["batch_size"] = int(ssplit[2][1:])
    d["seq_len"] = int(ssplit[3][1:])
    d["future_period_predict"] = int(ssplit[4][3:])
    d["time"] = int(ssplit[5])
    d["epoch"] = int(ssplit[6][1:])
    d["vacc"] = float(ssplit[7][4:])

    return d

for d in ls:
    formatted_ls.append(format_name(d))
```

```python
def sort_by_vacc(e):
    return e["vacc"]

def sort_by_time(e):
    return e["time"]
```

```python
formatted_ls = [d for d in formatted_ls if d["title"] == "softmax2"]

formatted_ls.sort(key=sort_by_vacc, reverse=True)

print(json.dumps(formatted_ls, indent=4))
```
